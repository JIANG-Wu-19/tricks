# 领票规则
每位同学可以在活动链接中填写一个1-1000的整数；
截止时间后计算所有同学填写数字的平均值，取其三分之二为参考值；
按照参与同学所填数字与参考值的接近程度排序取前50名；

# 建立数学模型
* 从一位同学开始模拟，随机生成i个1-1000的整数
* 根据上述公式计算参考值，并根据参考值进行填写

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
avg=[]
ind=[]
for i in tqdm(range(1,20001)):
    sum=0
    num=[]
    for j in range(1,i+1):
        tmp=random.randint(1,1000)
        num.append(tmp)
        sum+=tmp
    average=sum/i*2/3.0
    avg.append(average)
    ind.append(i)

avg=np.array(avg)
print(avg[-50:])
ind=np.array(ind)
plt.plot(ind,avg)
plt.show()
```
